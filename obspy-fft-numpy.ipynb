```python
from obspy.clients.fdsn import Client
from obspy import UTCDateTime
```

```python
client = Client('IRIS')
time = UTCDateTime('2021-10-07T13:53:23')
```

```python
starttime = time - 60
endtime = time +15*60
net = 'IU'
sta = 'MAJO'
loc = '10'
chan = 'BH*'
```

```python
st = client.get_waveforms(net, sta, loc, chan, starttime, endtime)
print(st)
st.plot()
```

```python
import matplotlib.pyplot as plt

st.spectrogram(wlen=50,log=True)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Data = []
for i in range(len(st)):
  Dataa = st[i].data
  Data.append(Dataa)
```

```python
Datafft = []
for i in range(len(Data)):
  datafft = np.fft.rfft(Data[i])
  Datafft.append(datafft)
  plt.figure(i+1)
  plt.plot(abs(Datafft[i]),linewidth = 0.5)
  plt.title(str(st[i]))
  plt.xscale('log')
  plt.xlabel('Freq (Hz)')
  plt.yscale('log')
  plt.ylabel('Amplitude')
  plt.grid()
```
